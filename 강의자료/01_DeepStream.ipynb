## 01-2. Computer Vision이란?
Computer Vision(CV)은 컴퓨터가 이미지를 이해하고 해석하도록 하는 기술입니다. 인간의 시각적 인식을 모방하여 이미지나 영상을 분석하고 의미 있는 정보를 추출하는 것을 목표로 합니다.

### 1. **Computer Vision의 기본 요소**
* 픽셀과 채널
    * 픽셀(Pixel): 디지털 이미지를 구성하는 가장 작은 단위.
    * 채널(Channel): RGB(빨강, 초록, 파랑)로 이루어진 컬러 이미지 또는 하나의 채널로 구성된 그레이스케일 이미지.
* 이미지 표현
    * 이미지는 2차원 배열(가로 x 세로 픽셀)로 표현되며, 각 픽셀에는 밝기 또는 색상 값이 저장됩니다.
    * 컬러 이미지에서는 각 픽셀이 RGB 값으로 표현됩니다.
* 해상도와 크기
    * 해상도: 이미지의 가로와 세로 픽셀 수를 의미.
    * 고해상도 이미지는 더 많은 디테일을 포함하지만 처리 시간이 더 오래 걸립니다.

### 2. **주요 기술과 기법**
* **이미지 분류(Image Classification)**: 이미지의 주요 내용을 특정 카테고리로 분류.
    * 예: 고양이 사진인지 개 사진인지 분류.
* **객체 검출(Object Detection)**: 이미지 내 특정 객체를 찾고, 위치를 Bounding Box로 표시.
    * 예: 자율주행에서 보행자와 차량 탐지.
* **세그멘테이션(Segmentation)**: 이미지의 각 픽셀을 특정 클래스에 할당.
    * Semantic Segmentation: 같은 클래스의 모든 객체를 하나로 처리.
    * Instance Segmentation: 동일한 클래스의 개별 객체를 구분.
* **Obhect Tracking**: 연속된 영상 프레임 간 픽셀 이동을 분석해 움직임 추적.
    * 예: 비디오 분석에서 객체 이동 경로 추적.

### 3. **활용 분야**
* 의료: X-ray, MRI, CT 스캔 이미지 분석.
* 자동차: 자율주행 시스템에서 도로, 차량, 보행자 인식.
* 보안: 얼굴 인식, CCTV 영상 분석.
* 소셜 미디어: 이미지 태깅, 필터 추천.
* 산업 자동화: 품질 검사, 로봇 비전.

## 01-3. DeepStream이란?
<img src="./img/01_DeepStream/DeepStream.png"><br>

DeepStream은 NVIDIA가 제공하는 AI 기반 비디오 분석 프레임워크입니다. GPU의 강력한 병렬 처리 능력을 활용하여 실시간으로 대규모 데이터를 처리하고, 딥러닝 모델을 실행할 수 있도록 설계되었습니다. 또한, DeepStream은 NVIDIA TensorRT, CUDA, 그리고 Triton Inference Server와 통합되어 높은 성능과 유연성을 제공합니다. DeepStream의 장점은 다음과 같습니다.

1. **실시간 성능**
DeepStream은 NVIDIA GPU의 병렬 처리를 최적화하여 초당 수천 개의 프레임을 처리할 수 있습니다. 이는 딥러닝 모델을 여러 스트림에 동시에 적용해야 하는 상황에서 큰 강점입니다.

2. **NVIDIA 그래픽 카드와의 호환성**
DeepStream은 NVIDIA의 그래픽 카드와 완벽히 호환되며, NVIDIA A100이나 H100 GPU와 같은 최신 하드웨어를 활용하면 대규모 비디오 분석 시스템을 쉽게 구축할 수 있습니다. 또한, CUDA와 TensorRT 같은 NVIDIA 소프트웨어 스택과도 최적화되어 있어, 하드웨어와 소프트웨어 간의 시너지를 극대화할 수 있습니다.

3. **TAO 및 NiM으로 최적화된 모델 전이 학습 지원**
DeepStream은 NVIDIA TAO(Train, Adapt, Optimize) 및 NiM(NVIDIA Inference Manager)을 활용하여 사전 학습된 딥러닝 모델을 간편하게 전이 학습할 수 있습니다. 이를 통해 사용자는 자신의 데이터셋에 맞는 맞춤형 모델을 빠르게 생성할 수 있습니다.

4. **플러그인 기반 아키텍처**
DeepStream은 플러그인 기반으로 설계되어 있습니다. 이를 통해 데이터 전처리, 모델 추론, 후처리와 같은 작업을 손쉽게 확장하고 사용자 정의할 수 있습니다.

5. **멀티모달 처리**
DeepStream은 비디오, 오디오, IoT 센서 데이터를 통합 분석할 수 있는 멀티모달 처리 기능을 지원합니다. 이로 인해 스마트 시티와 같은 복잡한 환경에서의 통합 분석이 가능합니다.

6. **메시지 큐 통합**
DeepStream은 Redis, MQTT, Kafka와 같은 메시지 큐 시스템과 통합하여 분석 결과를 실시간으로 전송하거나, 분산 환경에서 데이터를 효율적으로 관리할 수 있습니다. 이를 통해 스트림 처리와 이벤트 기반 데이터 파이프라인 구축이 간편해집니다.

7. **엣지 디바이스 지원**
Jetson Nano, Xavier와 같은 NVIDIA Jetson 디바이스와 완벽히 통합되어 에지에서 실시간 분석을 수행할 수 있습니다.

이러한 장점을 활용하여 DeepStream은 주로 다음과 같은 응용 분야에서 사용됩니다.

- 스마트 시티: 교통 감시 및 혼잡 관리
- 리테일: 고객 행동 분석 및 매장 최적화
- 헬스케어: 환자 모니터링 및 진단
- 산업 자동화: 품질 검사 및 로봇 비전

## 01-4. 컴퓨터 비전 프레임 워크 비교
### 1. OpenCV
<img width="10%" height="10%" src="https://img.icons8.com/color/480/opencv.png" alt="opencv"/>
<br />
OpenCV는 이미지 및 비디오 처리에 널리 사용되는 라이브러리입니다. 그러나 딥러닝 추론이나 대규모 비디오 스트림 분석에서는 DeepStream보다 성능이 낮습니다. OpenCV는 주로 간단한 이미지 처리나 시각화에 적합하며, DeepStream은 GPU 가속과 딥러닝 통합이 필요한 대규모 프로젝트에 적합합니다.

### 2. TensorFlow/ONNX Runtime
<img width="10%" height="10%" src="https://img.icons8.com/color/480/tensorflow.png" alt="tensorflow"/>
<br />
TensorFlow와 ONNX Runtime은 딥러닝 모델 추론에 강점을 가지고 있지만, 비디오 스트림 처리와 같은 특정 영역에서는 최적화가 부족합니다. DeepStream은 모델 추론뿐 아니라 스트림 관리, 전처리, 후처리를 포함한 전체 워크플로우를 지원합니다.

### 3. GStreamer

<br />
DeepStream은 GStreamer 기반으로 설계되었지만, 비디오 분석에 특화된 플러그인을 제공합니다. GStreamer는 멀티미디어 스트리밍에 강점이 있지만, 딥러닝 모델 추론이나 GPU 최적화는 DeepStream만큼 강력하지 않습니다.

### 4. 데이터 스트리밍 프레임워크
<img width="10%" height="10%" src="https://img.icons8.com/nolan/512/apache-kafka.png" alt="apache-kafka"/>
<img width="10%" height="10%" src="https://img.icons8.com/color/480/redis--v1.png" alt="redis--v1"/>
<br />
데이터 스트리밍 프레임워크인 Kafka, Redis 등과 같은 메세지큐는 데이터 스트리밍과 처리에 강점을 가지고 있지만, 비디오 프레임 분석과 같은 작업을 직접 처리하지 않습니다. DeepStream은 데이터 스트리밍 프레임워크와 통합하여 실시간 분석 결과를 스트리밍 하고, 데이터 파이프라인을 확장할 수 있습니다.

## 01-5. DeepStream 프로세스


DeepStream은 NVIDIA의 GStreamer 기반 멀티미디어 프레임워크를 확장하여 설계되었습니다. 이를 통해 다양한 입력 데이터 형식과 AI 모델을 결합한 실시간 스트리밍 분석이 가능합니다. 주요 구성 요소는 다음과 같습니다.

- Source : 비디오 데이터의 입력을 담당하며, 카메라 스트림, 파일, RTSP 등 다양한 입력 소스를 지원합니다.
- Decoder : 입력 데이터를 압축 해제하여 프레임 단위로 변환합니다.
- Pre-Processing : 영상 데이터를 AI 모델에 입력하기 전에 전처리 작업을 수행합니다.
- Inference : TensorRT를 통해 딥러닝 모델을 실행하여 객체 감지, 분류, 세그멘테이션 등 다양한 분석 작업을 수행합니다.
- Post-Processing : 모델의 추론 결과를 정리 및 시각화하여 사용자에게 전달합니다.
- Sink : 분석 결과를 출력하거나 다른 시스템으로 전송합니다. 출력 형식으로는 화면 디스플레이, 파일 저장, 메시지 큐 전송 등이 있습니다.

DeepStream은 블록 단위로 프로세스를 진행하며, 각 프로세스에 알맞는 conf파일을 적용하여 블록을 정의할 수 있습니다. 정의된 블록들은 DeepStream의 메인 코드에서 로직을 구현할 수 있습니다.

### Nouveau 비활성화
기본 내장된 nvidia 드라이버와 충돌을 막기 위해 Nouveau 비활성화
```bash
# nouveau 드라이버를 블랙리스트에 추가
$ sudo bash -c "echo blacklist nouveau > /etc/modprobe.d/blacklist-nvidia-nouveau.conf"
$ sudo bash -c "echo options nouveau modeset=0 >> /etc/modprobe.d/blacklist-nvidia-nouveau.conf"
 
# 아래 명령어로 blacklist-nvidia-nouveau.conf 파일 내용 확인
$ cat /etc/modprobe.d/blacklist-nvidia-nouveau.conf
blacklist nouveau
options nouveau modeset=0
 
# 아래 명령어 입력 후 재부팅
$ sudo update-initramfs -u
 
# nouveau 실행 확인 -> 아무것도 안뜨면 종료된 것
$ lsmod | grep nouveau
```

### DeepStream 필수 라이브러리 설치

#### glib 라이브러리 설치
```bash
# pip 설치
$ sudo apt install python3-pip

# pip을 통한 라이브러리 설치
$ sudo pip3 install meson
$ sudo pip3 install ninja

# glib 설치
$ git clone https://github.com/GNOME/glib.git
$ cd glib
$ git checkout 2.76.6
$ sudo meson build --prefix=/usr
$ sudo ninja -C build/
$ cd build/
$ sudo ninja install

# glib 확인
$ pkg-config --modversion glib-2.0
```

#### Kafka 라이브러리 설치
```bash
# Kafka 라이브러리 설치
$ git clone https://github.com/confluentinc/librdkafka.git
$ cd librdkafka
$ git checkout tags/v2.2.0
$ ./configure --enable-ssl
$ make
$ sudo make install

# Kafka 라이브러리를 DeepStream 라이브러리에 복사
$ sudo mkdir -p /opt/nvidia/deepstream/deepstream/lib
$ sudo cp /usr/local/lib/librdkafka* /opt/nvidia/deepstream/deepstream/lib
$ sudo ldconfig
```
